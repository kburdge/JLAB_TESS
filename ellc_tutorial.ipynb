```python
import numpy as np
import matplotlib.pyplot as plt
import ellc
```

## Generating a light curve with basic parameters

```python
# generate timestamps
t = np.arange(0, 50, 0.05)
```

```python
# set parameters for ELLC's lightcurve model
radius_1 = 0.3
radius_2 = 0.3
incl = 85
t_zero = 3
sbratio = 0.1
period = 20

model = ellc.lc(
    t_obs=t,
    radius_1=radius_1,
    radius_2=radius_2,
    incl=incl,
    t_zero=t_zero,
    sbratio=sbratio,
    period=period,
)
```

```python
# Plot the synthetically generated light curve.
# The plot shows the primary eclipses happening at t_zero and subsequent times offset by multiples of the period
# The graph is overall flat outside the eclipses-- this will change once we add in more parameters such as in the next section!
plt.scatter(t, model)
```

## Adding more parameters into the mix

```python
# Let's see how adding new parameters, heat_1 and heat_2, affect the light curve from the previous section.
radius_1 = 0.3
radius_2 = 0.3
incl = 85
t_zero = 0.5
sbratio = 10
period = 20
heat_1 = 10
heat_2 = 6

model = ellc.lc(
    t_obs=t,
    radius_1=radius_1,
    radius_2=radius_2,
    incl=incl,
    t_zero=t_zero,
    sbratio=sbratio,
    heat_1=heat_1,
    heat_2=heat_2,
    period=period,
)
```

```python
# Notice how the new heat parameters affect the curvature of the light curve.
# The sections outside the eclipses are no longer flat. This is due to the added effects of light from one star reflecting off its companion star, which
# are related to the heat coefficients.
plt.scatter(t, model)
```

```python
# We can also specify the shape parameters of the star. By default, stars are modeled as spherical, but setting the parameter
# to be 'roche' allows ELLC to describe star deformations from the gravitational forces that exist between the stars.
radius_1 = 0.3
radius_2 = 0.3
incl = 85
t_zero = 0.5
sbratio = 10
period = 20
heat_1 = 10
heat_2 = 6

model = ellc.lc(
    t_obs=t,
    radius_1=radius_1,
    radius_2=radius_2,
    incl=incl,
    t_zero=t_zero,
    sbratio=sbratio,
    shape_1="roche",
    shape_2="roche",
    period=period,
)
```

```python
plt.scatter(t, model)
```
